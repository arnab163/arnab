```python
!pip install nemoguardrails

import os

# Create the config directory if it doesn't exist
config_dir = "./config"
if not os.path.exists(config_dir):
    os.makedirs(config_dir)

# Create a minimal config.py inside the config directory
config_py_content = """
from nemoguardrails.llm.providers import register_llm_provider
from nemoguardrails.llms.llm import LLM

class DummyLLM(LLM):
    def __init__(self):
        super().__init__()
        self.model = "dummy"

    async def generate_async(self, prompt: str, stop=None, **kwargs):
        # For demonstration, this LLM returns a static dummy response.
        return "This is a dummy response from the LLM."

register_llm_provider("dummy", DummyLLM)

llm_configs = [
    {
        "type": "dummy",
        "name": "default_llm"
    }
]
"""
with open(os.path.join(config_dir, "config.py"), "w") as f:
    f.write(config_py_content)

# Create a minimal rail_spec.py inside the config directory
rail_spec_content = """
define user express greeting
  "hello"
  "hi"

define flow greeting
  user express greeting
  bot express greeting

define bot express greeting
  "Hello! How can I help you?"
"""
with open(os.path.join(config_dir, "rail_spec.py"), "w") as f:
    f.write(rail_spec_content)

from nemoguardrails import RailsConfig, LLMRails

config = RailsConfig.from_path(config_dir)
rails = LLMRails(config)

print("Guardrails loaded")
```

```python
import asyncio

async def run():
    response = await rails.generate_async(
        messages=[{"role": "user", "content": "Explain LLM in simple terms"}]
    )
    print(response)

await run()
```
